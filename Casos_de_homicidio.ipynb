## Analizando dados de pesquisas
A primeira analize foi usando os dados do IPAE, pegamos os arquivos `.csv` que se referiam a quantidade de homicidios ocorridos em todas as Unidades Federativas desde 1979 a 2017 e também usamos os dados de outro arquivo retirado do próprio IPAE que nos informa os homicidios totais no Brasil todo desde 1979 a 2017.

## Analize de todos os homicidios no Brasil entre 2007 a 2017


````
import pandas as pd
import requests
import matplotlib.pyplot as plt

dados_uf = pd.read_csv("homicidios_uf.csv", sep=';')
dados_brasil = pd.read_csv("homicidios_brasil.csv",sep=';')
dados_totais_uf = []
#dados_totais_uf = [{'nome':dados_uf[dados_uf.nome =="MG"]}]
for nome in dados_uf[dados_uf.período == 1979].nome:
  soma = sum(dados_uf[(dados_uf.nome == nome) & (dados_uf.período > 2006)].valor)
  dados_totais_uf.append({'nome':nome,'total':soma})

dataframe_dados_totais_uf = pd.DataFrame(dados_totais_uf)
dados_exame_completo = dados_brasil[dados_brasil.período >= 2007]
plt.plot(dados_exame_completo.período,dados_exame_completo.valor)
plt.xlabel("Ano")
plt.ylabel("Total de homicidios")
plt.grid(True)
plt.title("Total de homicidos no brasil entre 2007 a 2017")
plt.show()

plt.plot(dados_exame_completo.período,dados_exame_completo.valor,'k--', label="Total")
plt.plot(dados_exame_completo.período,dados_exame_completo.valor/100*91.8, label="Homens")
plt.plot(dados_exame_completo.período,dados_exame_completo.valor/100*8, label="Mulheres")
plt.xlabel("Ano")
plt.ylabel("Total de homicidios")
plt.grid(True)
plt.legend()
plt.title("Total de homicidos no brasil dividindo homens e mulheres 2007 a 2017")
plt.show()

plt.plot(dados_exame_completo.período,(((dados_exame_completo.valor/100)*91.8)/100)*15.9, label="Homens")
plt.plot(dados_exame_completo.período,(((dados_exame_completo.valor/100)*8)/100)*39.2, label="Mulheres")
plt.xlabel("Ano")
plt.ylabel("Total de homicidios")
plt.grid(True)
plt.legend()
plt.title("Total de vitimas de homicidos no brasil dividindo homens e mulheres de 2007 a 2017 a domicilio")
plt.show()
````
